## Market Size Analysis

* Market size analysis estimates the total size of the market for a specific product
  or service in terms of revenue, units sold or potential growth. It also identifies the 
  current market share of a product/service or company and also predicts how it may change in future.

* In the context of Electric Vehicles (EV's).It will involve Market Segmentation(which includes types of EV,
  fueling types), Competitive Analysis (market share of each company in term of 
  units sold or revenue), Consumer Preference and Future Growth.
  
The provided dataset contains the following columns, each representing different aspects of the electric vehicle (EV) population in the United States:

* **VIN (1-10):** Partial Vehicle Identification Number.
* **County:** The county in which the vehicle is registered.
* **City:** The city in which the vehicle is registered.
* **State:** The state in which the vehicle is registered. It appears that this dataset may be focused on Washington (WA) state.
* **Postal Code:** The postal code where the vehicle is registered.
* **Model Year:** The year of the vehicle model.
* **Make:** The manufacturer of the vehicle.
* **Model:** The model of the vehicle.
* **Electric Vehicle Type:** The type of electric vehicle, e.g., Battery Electric Vehicle (BEV).
* **Clean Alternative Fuel Vehicle (CAFV) Eligibility:** Eligibility status for clean alternative fuel vehicle programs.
* **Electric Range:** The maximum range of the vehicle on a single charge (in miles).
* **Base MSRP:** The Manufacturer’s Suggested Retail Price.
* **Legislative District:** The legislative district where the vehicle is registered.
* **DOL Vehicle ID:** Department of Licensing Vehicle Identification.
* **Vehicle Location**: Geographic coordinates of the vehicle location.
* **Electric Utility:** The electric utility service provider for the vehicle’s location.
* **2020 Census Tract:** The census tract for the vehicle’s location.

### We will begin by identifying the most popular manufacturers and their respective models within the electric vehicle (EV) market.

```python
# Visualizing popular EV manufacturers
plt.figure(figsize=(12, 6))
sns.countplot(y='Make', data=df, order=df['Make'].value_counts().head(15).index)
plt.title('Top 15 Popular EV Manufacturers')
plt.xlabel('No of Vehicles Sold by Manufacturers')
plt.show()
```

**The graph illustrates the top 15 most popular electric vehicle (EV) manufacturing companies.**

* **Tesla** leads by a significant margin, with approximately 80,000 vehicles sold, making it the most popular EV manufacturer.
* **Nissan and Chevrolet** follow in second and third place, respectively, with comparable sales figures, though significantly       lower than Tesla's.
* The remaining manufacturers **Ford, BMW, KIA, Toyota, Volkswagen, Jeep, Hyundai, Rivian, Volvo, Audi, Chrysler, and Mercedes-     Benz** report lower sales compared to the top three.

```python
# Top manufacturers
top_ev_manufacturers = df['Make'].value_counts().head(15)

top_models = (
    df[df['Make'].isin(top_ev_manufacturers.head(5).index)]  # Filtering for top 5 manufacturers
    .groupby(['Make', 'Model'])  # Grouping by Make and Model
    .size()  # Count occurrences
    .reset_index(name='No of Models registered')  # Converting to DataFrame
    .sort_values(by='No of Models registered', ascending=False)  # Sorting by count
    .head(15)  # Selecting top 15 models
)
                
# Visualizing the top 15 models
plt.figure(figsize=(12, 6))
sns.barplot(data=top_models, x='No of Models registered', y='Model', hue='Make', palette='viridis')
plt.title('Top 10 EV Models Across Top 5 Manufacturers')
plt.xlabel('No of Models registered')
plt.ylabel('Model')
plt.legend(title='Manufacturer', loc='lower right')
plt.tight_layout()
plt.show()
```

**The graph above illustrates the top 15 most popular models from the top 5 manufacturers.**

* **Tesla's Model Y and Model 3** dominate, with a combined total of over 65,000 registrations.
* The **Nissan Leaf** emerges as the third most popular model and the first non-Tesla entry on the list.
* Following the top three models, the remaining vehicles, in decreasing order of popularity, include **Model S, Leaf, Bolt EV,     Model X, Volt, Mustang Mach-E, X5, i3, Fusion, Bolt EV, F-150, C-MAX, and i4**.

##  Electric Vehicle Adoption Trends

```python
EV_adoption = df['Model Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(EV_adoption.index, EV_adoption.values, color='purple')
plt.title('EV Adoption Trends Over The Years')
plt.xlabel('Model Year')
plt.ylabel('Number of Vehicles')
plt.tight_layout()
plt.show()
```

**The graph above illustrates EV adoption trends over time.**
* The dataset spans from 1997 to 2024, though data for 2024 is incomplete.
* EV adoption began to accelerate significantly from 2011 onwards, with 2023 recording the highest number of EV registrations     within the period.

```python
# Visualizing distribution of Electric Vehicle Types
Ev_type = df['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=Ev_type.index, y=Ev_type.values, palette='viridis')
plt.title('Distribution of Electric Vehicle Types')
plt.xlabel('Electric Vehicle Type')
plt.ylabel('Count')
plt.show()
```

**The graph above illustrates the distribution of Electric Vehicle (EV) types in the dataset.**
* The dataset includes two EV types: **Battery Electric Vehicles (BEVs) and Plug-in Hybrid Electric Vehicles (PHEVs).**
* BEVs have a significantly higher number of registrations compared to PHEVs, indicating greater demand for BEVs.

```python
# Visualizing distribution of Electric Vehicle Range 
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='Electric Range', bins=30, color='green', kde=True)
plt.title('Distribution of EV Range (In Miles)')
plt.xlabel('Electric Range')
plt.ylabel('Count')
plt.show()
```

**The graph above illustrates the distribution of EV ranges (in miles).**
* The EV range varies from 10 to 350 miles.
* A majority of EVs have a range between 10 and 50 miles.

```python
# Distribution of Electric Range 
top_5_ev_make_counts = df['Make'].value_counts().head(5).index

# Filtering for top 5 makes
df_top_5_makes = df[df['Make'].isin(top_5_ev_make_counts)]

# Getting counts for each model within the top 5 makes
top_10_models = df_top_5_makes['Model'].value_counts().head(10).index

# Filtering top 15 models from top 5 makes
df_top_10_models = df_top_5_makes[df_top_5_makes['Model'].isin(top_10_models)]

# Calculating range statistics for each model
range_stats = (
    df_top_10_models.groupby('Model')['Electric Range']
    .agg(['min', 'max', 'mean'])
    .rename(columns={'min': 'Min Range', 'max': 'Max Range', 'mean': 'Mean Range'})
    .reset_index()
)

# Merging range statistics with the filtered dataframe
range_stats_with_make = pd.merge(df_top_10_models[['Model', 'Make']], range_stats, on='Model', how='left')

# Setting up figure and axes
fig, axes = plt.subplots(1, 2, figsize=(50, 18)) # 1 row, 2 columns of plots

# Bar plot for Mean Range
sns.barplot(x='Mean Range', y='Model', data=range_stats_with_make, ax=axes[0], palette='viridis')
axes[0].set_title('Mean Electric Range for Top 15 Models')
axes[0].set_xlabel('Mean Range (miles)')
axes[0].set_ylabel('Model')
axes[0].tick_params(axis='y', labelsize=10)

# Bar plot for Min and Max range
range_stats_melted = range_stats_with_make.melt(id_vars=['Model'], value_vars=['Min Range', 'Max Range'], var_name = 'Range Type', value_name = 'Range')
sns.barplot(x='Range', y='Model', hue='Range Type', data=range_stats_melted, ax=axes[1], palette = ['lightcoral', 'skyblue'])
axes[1].set_title('Min and Max Electric Range for Top 15 Models')
axes[1].set_xlabel('Range (miles)')
axes[1].set_ylabel('Model')
axes[1].tick_params(axis='y', labelsize=10)

# Adjusting layout and display plot
# plt.tight_layout()
plt.show()
```

```python
# Geographical Distribution
ev_county = df['County'].value_counts().head(5).index

# Analyzing the distribution of EVs within the cities of these top counties
top_cities = (
     df[df['County'].isin(ev_county)]  # Filtering for top counties
    .groupby(['County', 'City'])  # Grouping by County and City
    .size()  # Counting occurrences
    .reset_index(name='Number of Vehicles')  # Converting to DataFrame
    .sort_values(by='Number of Vehicles', ascending=False)  # Sorting by count
    .head(15)  # Selecting top 10 cities
)

# Visualizing the top 15 cities across these counties
plt.figure(figsize=(12, 8))
sns.barplot(data=top_cities, x='Number of Vehicles', y='City', hue='County', palette='magma')
plt.title('Top Cities in Top Counties by EV Registrations')
plt.xlabel('Number of Vehicles Sold')
plt.ylabel('City')
plt.legend(title='County', loc='lower right')
plt.tight_layout()
plt.show()
```

**The graph illustrates the geographical distribution of EV registrations within Washington DC State.**

* **King** County has the highest number of EV registrations.
* Following **King County** are **Clark and Snohomish**, with **Thurston and Pierce** rounding out the top five counties.
* The top three cities, all located within **King County**, are **Seattle, Bellevue, and Redmond.**
  **Everett**, in **Snohomish County**, has the lowest number of EV registrations.
